準備

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

定義どおり実装する

平均変化率

```python
def Average_change_rate(f,a,b):
    delta_x = b - a
    delta_y = f(b)-f(a)
    return delta_y / delta_x


def f(x):
    return x ** 2


a= 1
b = 4

print("平均変化率：", Average_change_rate(f, a, b))

x = np.arange(-3.0, 5.0, 0.1)

y = f(x)
d = Average_change_rate(f, a, b)
y2 = d * x + f(a) - d*a

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

微分

```python
def df(f, x, h):
    return (f(x+h) - f(x)) / (h)


def f(x):
    return x ** 2

x = np.arange(-3.0, 5.0, 0.1)
h = 1e-4
a = 1

y = f(x)
d = df(f, a, h)
y2 = f(a) + d * (x - a)

print("微分係数：", d)

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
