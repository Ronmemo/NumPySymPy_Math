準備

```python
import sympy as sym
from sympy.plotting import plot
sym.init_printing(use_unicode=True)
```

変数の定義

```python
# 変数の定義
x = sym.symbols("x")
```

一次関数

```python
y = 2*x+1
plot(y)
```

二次関数

```python
y = x**2+4*x+4
plot(y)
```

三角関数

```python
y1 = sym.cos(x)
y2 = sym.sin(x)
plot(y1, y2, (x, -2*sym.pi, 2*sym.pi))
```

指数関数と対数関数

```python
y1 = sym.exp(x)
y2 = sym.log(x)
plot((y1, (x, -2.00, 2.00)), (y2, (x, 0.01, 4.00)))
```
