準備

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

一次関数

```python
x = np.arange(-10, 10, 0.1)
y = 2*x+1

plt.plot(x, y)
plt.show()
```

二次関数

```python
x = np.arange(-10, 10, 0.1)
y = x**2+4*x+4

plt.plot(x, y)
plt.show()
```

三角関数

```python
x = np.arange(0, 2*np.pi, 0.01)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1, label="cos x", color="red")
plt.plot(x, y2, label="sin x", color="blue")
plt.legend()
plt.show()
```

指数関数と対数関数

```python
x1 = np.arange(-2.00, 2.00, 0.01)
y1 = np.exp(x1)

x2 = np.arange(0.01, 4.00, 0.01)
y2 = np.log(x2)

plt.plot(x1, y1, label="exp", color="red")
plt.plot(x2, y2, label="log", color="blue")
plt.legend()
plt.show()
```
