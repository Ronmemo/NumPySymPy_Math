準備

```python
import sympy as sym
from sympy.plotting import plot
sym.init_printing(use_unicode=True)

# 変数の定義
x = sym.symbols("x")
```

例：二次関数の極限

```python
y = x**2+4*x+4
plot(y)
```

```python
#xを0に近づける，(関数，対象の変数，変数を何に近づけるか)
sym.limit(y, x, 0)
```

```python
#xを正の無限大に近づける
sym.limit(y, x, sym.oo)
```

例：三角関数の極限で有名な公式

```python
y = sym.sin(x)/x
plot(y)
```

```python
#この関数は0に代入では定義できない
y.subs(x, 0)
```

```python
#xを0に近づけると求まる
sym.limit(y, x, 0)
```

例：三角関数の極限

```python
y = sym.sin(x)
plot(y)
```

```python
#xを正の無限大に近づける
sym.limit(y, x, sym.oo)
```
